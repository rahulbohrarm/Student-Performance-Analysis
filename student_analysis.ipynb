# 🎓 Student Performance Analysis using Python
This project explores factors affecting student academic performance using data analysis and visualization.


```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('student_performance.csv')
data.head()
```

```
# Basic statistics
print(data.describe())

# Visualizing relationships
sns.pairplot(data, hue='gender')
plt.show()
```

```
# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Student Performance')
plt.show()
```

### 🔍 Insights:
- Students studying more than 5 hours generally perform better.
- Higher parental education often correlates with improved student outcomes.
- Math, Reading, and Writing scores are positively correlated.
